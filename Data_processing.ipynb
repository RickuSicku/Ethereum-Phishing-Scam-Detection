```python
from_count = df['from'].value_counts()
to_count = df['to'].value_counts()
```

```python
selected_columns1 = df[["from","fromIsPhi"]]
new_df1 = selected_columns1.copy()
selected_columns2 = df[["to","toIsPhi"]]
new_df2 = selected_columns2.copy()
new_df1.rename(columns = {'from':'ID', 'fromIsPhi':'IsPhi'}, inplace = True)
new_df2.rename(columns = {'to':'ID', 'toIsPhi':'IsPhi'}, inplace = True)
dataframe = [new_df1, new_df2]
allids = pd.concat(dataframe)
uniqueids2=allids.drop_duplicates()
uniqueids2
```

```python
# prompt: for ID in uniqueids2 , find the count of each ID in from and to of dataframe df , and add ID , count_from and count_to in a new single dataframe

import pandas as pd
id_counts = []
for ID in uniqueids2['ID']:
  from_count = df[df['from'] == ID].shape[0]
  to_count = df[df['to'] == ID].shape[0]
  id_counts.append({'ID': ID, 'count_from': from_count, 'count_to': to_count})

id_counts_df = pd.DataFrame(id_counts)
```

```python
account_mapping = {account_id: idx for idx, account_id in enumerate(set(df['from'].unique()) | set(df['to'].unique()))}
```

```python
node_features = torch.tensor(attributes[['avg_amount', 'active_days', 'avg_frequency','count_from','count_to']].values, dtype=torch.float)
target_values = torch.tensor(attributes['IsPhi'].values, dtype=torch.float)  # Replace 'some_target_column' with the actual target column in 'attributes'
edges_from = torch.tensor(df['from'].map(account_mapping).values, dtype=torch.long)
edges_to = torch.tensor(df['to'].map(account_mapping).values, dtype=torch.long)
edge_features = torch.tensor(df['amount'].values, dtype=torch.float).view(-1, 1)  # You can modify this based on your edge features
```

```python
data = Data(x=node_features, edge_index=torch.stack([edges_from, edges_to]), edge_attr=edge_features, y=target_values)
```
