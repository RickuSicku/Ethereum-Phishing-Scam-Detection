```python
import numpy as np
from google.colab import autoviz

def heatmap(df, x_colname, y_colname, figscale=1, mpl_palette_name='viridis'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  import pandas as pd
  plt.subplots(figsize=(8 * figscale, 8 * figscale))
  df_2dhist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sns.heatmap(df_2dhist, cmap=mpl_palette_name)
  plt.xlabel(x_colname)
  plt.ylabel(y_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = heatmap(_df_9, *['to', 'date'], **{})
chart
```

```python
condition = df['amount'] != 0

# Use boolean indexing to keep rows that meet the condition
df = df[condition]

print(df)
```

```python
df = df.sort_values(by='timestamp')
```

```python
# Define the size of each block (adjust as needed)
block_size = 10000  # Example: Each block contains 10,000 transactions

# Split the data into blocks
blocks = [df[i:i + block_size] for i in range(0, len(df), block_size)]
```

```python
import networkx as nx

# Create a list to store graphs for each block
graphs = []

for block in blocks:
    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes (accounts) to the graph
    G.add_nodes_from(block['from'])
    G.add_nodes_from(block['to'])

    # Add edges (transactions) to the graph
    edges = [(row['from'], row['to']) for _, row in block.iterrows()]
    G.add_edges_from(edges)

    # Store the graph in the list
    graphs.append(G)
```
